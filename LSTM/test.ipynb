```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import SGDRegressor
import pandas as pd

df = pd.read_csv('/home/gero/Documents/Austral/Redes Neuronales/Clase 2/automobile-simple.csv')

# Filtrar solo las columnas numéricas
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Reemplazar valores NaN con la media de cada columna
numeric_df.fillna(numeric_df.mean(), inplace=True)

# Normalizar los valores numéricos entre 0 y 1
scaler = MinMaxScaler()
numeric_df_scaled = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)

# Separar las variables independientes (X) y la variable dependiente (y)
X = numeric_df_scaled.drop('price', axis=1)
y = numeric_df_scaled['price']

# Crear y entrenar el modelo de regresión lineal con descenso de gradiente estocástico
sgd_regressor = SGDRegressor(max_iter=30, tol=10e-06, eta0=0.01)
sgd_regressor.fit(X, y)

# Verificar si el modelo ha convergido
has_converged = sgd_regressor.n_iter_ < 30

# Comprobar si la normalización es necesaria para la convergencia del algoritmo de entrenamiento
# Si el modelo ha convergido, entonces la afirmación es falsa; la normalización no es necesaria para que el algoritmo funcione
not has_converged
```

```python
# Realizar 50 entrenamientos independientes. Promediando los 50 vectores de pesos W se identificaron algunos atributos cuyo incremento disminuye el valor del precio.

# Crear un dataframe para almacenar los pesos W de cada entrenamiento
weights_df = pd.DataFrame(columns=X.columns)

# Realizar 50 entrenamientos independientes
for i in range(50):
    sgd_regressor = SGDRegressor(max_iter=30, tol=10e-06, eta0=0.01)
    sgd_regressor.fit(X, y)
    weights_df.loc[i] = sgd_regressor.coef_

# Calcular el promedio de los pesos W de los 50 entrenamientos
weights_df.mean()
```
