```python
import pandas as pd
from empresa4.core import get_dataset
```

```python
df = get_dataset("02_productos_todos_anti_leak")
df
```

```python
df.isna().sum()
```

```python
# from df, make a new df where periodo are the row headers and each column is a value od product_id. In each cell put the value of the "tn" column for that product_id and periodo
new_df = df.pivot(index='product_id', columns='periodo', values='tn')
new_df
```

```python
new_df.isna().sum()
```

```python
new_df.index
```

```python
# # Perform the transformation
# for col in new_df.columns[1:]:  # Skip 'Product' and 'Current Month' columns
#     new_df[col] = new_df[col] / new_df[201902]


# print(new_df)
```

```python
def custom_norm(row, col, date):
    # (x - x.min()) / (x.max() - x.min()), axis=1)
    my_min = row.min()
    my_max = row.max()
    if my_max == my_min:
        return 0
    return (row[col] - my_min) / (my_max - my_min)

for col in new_df.columns:  # Skip 'Product' and 'Current Month' columns
    new_df[f"{col}_norm"] = new_df.apply(custom_norm, axis=1,  args=([col, 201902]))

new_df
```

```python
new_df.isna().sum()
```

```python
final_df = new_df.drop(new_df.columns[0:26], axis=1)
final_df
```

```python
final_df.isna().sum()
```

```python
# apply umap to the final_df
import umap
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=6, random_state=42)
embedding = reducer.fit_transform(final_df)
embedding.shape
```

```python
# cluster the embedding
from sklearn.cluster import KMeans

# for values in range 2-10, calculate the inertia and plot it
inertia = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto').fit(embedding)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 10), inertia)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42).fit(embedding)

# report the cluster sizes
pd.Series(kmeans.labels_).value_counts()
```

```python
# apply umap to the final_df again but to two dimensions and plot the clusters
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=3, random_state=42)
embedding = reducer.fit_transform(final_df)
embedding.shape
```

```python
from mpl_toolkits.mplot3d import Axes3D

# plot the clusters in 3d
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(embedding[:, 0], embedding[:, 1], embedding[:, 2], c=kmeans.labels_, cmap='viridis', alpha=0.5)

plt.show()
```

```python
# apend the cluster label to the final_df
final_df['cluster'] = kmeans.labels_
final_df['cluster'].value_counts()
```
